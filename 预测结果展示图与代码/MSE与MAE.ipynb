```python
import pandas as pd
import matplotlib.pyplot as plt

# 指定文件路径
file_path = 'D:/Algorithm_Project/笔试题目/model/model/更改时间-20epochs-Interval-ETTh1.csv'

# 使用 Pandas 读取 CSV 文件
df = pd.read_csv(file_path)

# 每张子图显示的数据点数
chunk_size = 5000

# 计算需要绘制的子图数量
num_chunks = len(df) // chunk_size + 1

# 循环绘制每个子图
for i in range(num_chunks):
    start_idx = i * chunk_size
    end_idx = min((i + 1) * chunk_size, len(df))
    
    # 提取MSE和MAE数据
    mse_values = df['MSE'].iloc[start_idx:end_idx]
    mae_values = df['MAE'].iloc[start_idx:end_idx]

    # 创建新的图形
    plt.figure(figsize=(10, 6),dpi=300)

    # 绘制MSE和MAE数据
    plt.plot(mse_values, label='MSE', color='blue')
    plt.plot(mae_values, label='MAE', color='orange')

    # 添加标题和标签
    plt.title(f'Comparison of MSE and MAE (Chunk {i+1})', fontsize=16)
    plt.xlabel('Index', fontsize=14)
    plt.ylabel('Value', fontsize=14)

    # 设置纵轴最大刻度为70
    plt.ylim(0, 70)

    # 添加图例
    plt.legend()

    # 显示图形
    plt.show()
```
